```python
import csv
import re

def clean_text(text):
    # Hapus URL
    text = re.sub(r'http\S+', '', text)

    # Hapus karakter khusus, tanda baca, dan simbol lainnya
    text = re.sub(r'[^\w\s]', '', text)

    # Hapus emotikon
    text = re.sub(r'[\U0001F600-\U0001F64F\U0001F300-\U0001F5FF\U0001F680-\U0001F6FF\U0001F700-\U0001F77F\U0001F780-\U0001F7FF\U0001F800-\U0001F8FF\U0001F900-\U0001F9FF\U0001FA00-\U0001FA6F\U0001FA70-\U0001FAFF\U00002702-\U000027B0\U000024C2-\U0001F251]+', '', text)

    # Hapus karakter berulang, misalnya "hellooooo"
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Mengubah teks menjadi lowercase
    text = text.lower()

    return text

input_filename = '/content/drive/MyDrive/ùìºùì¥ùìªùì≤ùìπùìºùì≤ /dataset/segari_data_not_clean.csv'   # Nama file CSV input
output_filename = 'segari_dataset_file_bersih2.csv.csv' # Nama file CSV output

# Membaca file input dan menyimpan data dalam list
data = []
with open(input_filename, 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if len(row) > 0:
            data.append(row)

# Membersihkan teks pada setiap baris data
for row in data:
    if len(row) > 1:
        row[1] = clean_text(row[1])  # Menganggap kolom teks berada di indeks 1, sesuaikan jika berbeda

# Menyimpan data yang sudah dibersihkan ke dalam file output
with open(output_filename, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    for row in data:
        writer.writerow(row)
```
